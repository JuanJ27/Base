# Análisis de Actividad Radiactiva Utilizando Distribución de Poisson

## Introducción

La actividad radiactiva es un fenómeno natural caracterizado por la desintegración aleatoria de átomos inestables. Este proceso se rige por principios probabilísticos, donde un gran número de átomos tienen la posibilidad de desintegrarse en un intervalo de tiempo dado, aunque la probabilidad individual de desintegración de cada átomo es baja. Debido a esta naturaleza aleatoria y a la gran cantidad de átomos involucrados, la estadística de la actividad radiactiva se modela eficazmente mediante la **distribución de Poisson**.

El objetivo de este análisis es:

1. **Medir el fondo de radiación** utilizando un tubo escintilador.
2. **Determinar si las mediciones** realizadas siguen una distribución de Poisson.
3. **Evaluar el impacto** de la presencia de objetos adicionales, como bananos y muestras, en la distribución de la radiación.

## Datos

Se realizaron tres conjuntos de mediciones con el objetivo de analizar la actividad radiactiva bajo diferentes condiciones:

- **Sin Nada**: Medición del fondo de radiación sin la presencia de objetos adicionales.
- **Con Bananos**: Medición de la radiación con bananos colocados cerca del tubo escintilador.
- **Con Muestras**: Medición de la radiación con muestras específicas colocadas cerca del tubo escintilador.

### Conjuntos de Datos

- **Medida Sin Nada**: `[6, 4, 2, 3, 3, 3, 2, 3, 3, 2, 2, 1, 5, 4, 2, 7, 2, 2, 5, 2, 3, 4, 4, 2, 0, 1, 3, 6, 2, 0]`
- **Medida Con Bananos**: `[3, 6, 6, 4, 1, 4, 2, 2, 1, 5, 4, 1, 3, 5, 5, 1, 4, 2, 3, 1, 3, 5, 5, 6, 3, 6, 6, 5, 8, 2]`
- **Medida Con Muestras**: `[4, 3, 4, 3, 3, 5, 4, 3, 4, 7, 2, 6, 2, 3, 2, 4, 4, 3, 2, 1, 2, 4, 2, 1, 3, 6, 3, 4, 4, 3]`

## Metodología

El análisis se realizó siguiendo una serie de pasos estructurados para evaluar la adecuación de la distribución de Poisson a los datos obtenidos y para comparar las diferentes condiciones de medición.

### 1. Ajuste de la Distribución de Poisson a los Datos "Sin Nada"

El primer paso consistió en ajustar una distribución de Poisson a los datos obtenidos sin la presencia de objetos adicionales ("Sin Nada"). Para ello, se calculó el parámetro \(\lambda\), que representa el valor medio de los conteos de radiación. Posteriormente, se compararon las frecuencias observadas con las frecuencias esperadas según la distribución de Poisson utilizando la prueba de **chi-cuadrado** para evaluar la bondad del ajuste.

### 2. Comparación de "Con Bananos" y "Con Muestras" con la Distribución de Poisson de "Sin Nada"

Una vez establecido el modelo de Poisson basado en los datos "Sin Nada", se procedió a comparar las distribuciones de los datos "Con Bananos" y "Con Muestras" con este modelo. Esta comparación también se realizó mediante la prueba de chi-cuadrado para determinar si la presencia de bananos o muestras alteraba significativamente la distribución de la radiación en comparación con el fondo sin nada.

## Resultados

### 1. Ajuste de Poisson a "Sin Nada"

- **Valor Medio (\(\lambda\))**: 2.93
- **Desviación Estándar**: \(\sqrt{2.93} \approx 1.71\)
- **Prueba de Chi-cuadrado (\(\chi^2\))**: 4.08
- **Grados de Libertad**: 6
- **Valor Crítico (\(\chi^2_{crítico}\)) al 5%**: 12.592
- **Resultado de la Prueba**: **Sí** se acepta el ajuste a Poisson (4.08 < 12.592)

**Interpretación**:

Los datos "Sin Nada" siguen adecuadamente una distribución de Poisson, ya que el valor de chi-cuadrado es menor que el valor crítico al 5% de confianza. Esto indica que la distribución observada de los conteos de radiación se alinea bien con la esperada bajo una distribución de Poisson.

### 2. Comparación de "Con Bananos" con la Distribución de Poisson de "Sin Nada"

- **Prueba de Chi-cuadrado (\(\chi^2\))**: 6.42
- **Grados de Libertad**: 2
- **Valor Crítico (\(\chi^2_{crítico}\)) al 5%**: 5.991
- **Resultado de la Prueba**: **No** se acepta que "Con Bananos" siga la distribución de Poisson de "Sin Nada" (6.42 > 5.991)

**Interpretación**:

La presencia de bananos altera significativamente la distribución de radiación, alejándola de la distribución de Poisson observada en el fondo sin nada. Esto sugiere que los bananos afectan la tasa de detección de radiación, ya sea a través de la absorción, reflexión o emisión adicional de radiación.

- **Grados de Libertad**: 2
- **Valor Crítico (\(\chi^2_{crítico}\)) al 5%**: 5.991
- **Resultado de la Prueba**: **No** se acepta que "Con Muestras" siga la distribución de Poisson de "Sin Nada" (7.31 > 5.991)

**Interpretación**:

La presencia de muestras también altera significativamente la distribución de radiación, distorsionando la distribución de Poisson observada en el fondo sin nada. Esto indica que las muestras influyen de manera notable en la detección o emisión de radiación.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, chi2
import pandas as pd

# Datos proporcionados
tiempo=[3.25,2.10,4.93,11.08,2.60,1.28,5.61,5.49,1.10,5.52]
medida_sin_nada=[6,4,2,3,3,3,2,3,3,2,2,1,5,4,2,7,2,2,5,2,3,4,4,2,0,1,3,6,2,0]
medida_con_bananos=[3,6,6,4,1,4,2,2,1,5,4,1,3,5,5,1,4,2,3,1,3,5,5,6,3,6,6,5,8,2]
medida_con_muestras=[4,3,4,3,3,5,4,3,4,7,2,6,2,3,2,4,4,3,2,1,2,4,2,1,3,6,3,4,4,3]
print('Se realizaron',len(tiempo),'medidad', 'con una media de', round(np.mean(tiempo),2), 'segundos y desviacion estandar de', round(np.std(tiempo),1), 'segundos')
```

```python
def ajustar_poisson_sin_nada(datos, etiqueta='Sin Nada'):
    # Contar la frecuencia de cada conteo
    conteos, bins = np.histogram(datos, bins=range(min(datos), max(datos)+2), density=False)
    k = np.arange(min(datos), max(datos)+1)
    frecuencia_observada = conteos
    total = len(datos)
    
    # Calcular lambda (valor medio)
    lambda_est = np.mean(datos)
    
    # Calcular frecuencia esperada usando Poisson
    frecuencia_esperada = poisson.pmf(k, lambda_est) * total
            
    # Calcular chi-cuadrado
    chi_sq = np.sum((np.array(frecuencia_observada) - np.array(frecuencia_esperada))**2 / np.array(frecuencia_esperada))
    grados_libertad = len(frecuencia_observada) - 1 - 1  # bins - 1  parámetros estimado
    
    # Valor crítico de chi-cuadrado al 5%
    chi_crit = chi2.ppf(0.95, grados_libertad)
    
    # Decisión
    ajuste_bueno = chi_sq < chi_crit
    
    # Mostrar resultados
    print(f"--- {etiqueta} ---")
    print(f"Lambda estimado: {lambda_est:.2f}")
    print(f"Desviación estándar: {np.sqrt(lambda_est):.2f}")
    print(f"Grados de libertad: {grados_libertad}")
    print(f'El valor de chi cuadrado es {chi_sq:.2f} y el valor critico al 5% es {chi_crit:.2f}\nPor lo que el ajuste es {"bueno" if ajuste_bueno else "malo"}')
    
    # Plotear el histograma y la distribución de Poisson
    plt.figure(figsize=(8,5))
    plt.bar(k, frecuencia_observada, width=0.6, alpha=0.6, label='Observado', color='skyblue', edgecolor='black')
    plt.plot(k, frecuencia_esperada, 'ro-', label='Poisson Ajustado')
    plt.xlabel('Conteos')
    plt.ylabel('Frecuencia')
    plt.title(f'Histograma y Ajuste de Poisson - {etiqueta}')
    plt.legend()
    plt.show()
    
    return lambda_est, frecuencia_esperada, frecuencia_observada

def plot_histograma_con_ajuste(datos, lambda_ref, etiqueta='Con Bananos'):
    # Contar la frecuencia de cada conteo
    conteos, bins = np.histogram(datos, bins=range(min(datos), max(datos)+2), density=False)
    k = np.arange(min(datos), max(datos)+1)
    frecuencia_observada = conteos
    total = len(datos)
    
    # Calcular frecuencia esperada usando Poisson con lambda_ref
    frecuencia_esperada = poisson.pmf(k, lambda_ref) * total
    
    # Calcular chi-cuadrado
    chi_sq = np.sum((np.array(frecuencia_observada) - np.array(frecuencia_esperada))**2 / np.array(frecuencia_esperada))
    grados_libertad = len(frecuencia_observada) - 1 - 0  # bins -1 - 0 parámetros estimados (lambda está fijo)
    
    # Valor crítico de chi-cuadrado al 5%
    chi_crit = chi2.ppf(0.95, grados_libertad)
    
    # Decisión
    ajuste_bueno = chi_sq < chi_crit
    
    # Mostrar resultados
    print(f"--- {etiqueta} ---")
    print(f"Grados de libertad: {grados_libertad}")
    print(f'El valor de chi cuadrado es {chi_sq:.2f} y el valor critico al 5% es {chi_crit:.2f}\nPor lo que el ajuste es {"bueno" if ajuste_bueno else "malo"}')
    
    # Plotear el histograma y la distribución de Poisson de 'Sin Nada'
    plt.figure(figsize=(8,5))
    plt.bar(k, frecuencia_observada, width=0.6, alpha=0.6, label='Observado', color='lightgreen', edgecolor='black')
    frecuencia_esperada_plot = poisson.pmf(k, lambda_ref) * total
    plt.plot(k, frecuencia_esperada_plot, 'ro-', label='Poisson de Sin Nada')
    plt.xlabel('Conteos')
    plt.ylabel('Frecuencia')
    plt.title(f'Histograma y Poisson de Sin Nada - {etiqueta}')
    plt.legend()
    plt.show()
    
    return chi_sq, frecuencia_esperada_plot, frecuencia_observada

# Ajustar 'Sin Nada' a Poisson
lambda_sin_nada, freq_espe_nada, freq_obs_nada = ajustar_poisson_sin_nada(medida_sin_nada, 'Sin Nada')

# Comparar 'Con Bananos' con Poisson de 'Sin Nada'
chi_bananos, freq_espe_ban, freq_obs_ban  = plot_histograma_con_ajuste(medida_con_bananos, lambda_sin_nada, 'Con Bananos')

# Comparar 'Con Muestras' con Poisson de 'Sin Nada'
chi_muestras, freq_espe_mu, freq_obs_mu  = plot_histograma_con_ajuste(medida_con_muestras, lambda_sin_nada, 'Con Muestras')
```

```python
def grafica_de_residuos(y, ya):
    """
    Funcion que grafica los residuos de un ajuste de las distribuciones
    """
    plt.figure(figsize=(8,5))
    residuos = y - ya
    plt.plot(residuos, 'o-')
    plt.axhline(0, color='red', linestyle='--')
    plt.xlabel('Conteos')
    plt.ylabel('Residuos')
    plt.title('Residuos del ajuste')
    plt.show()

grafica_de_residuos([freq_obs_nada, freq_espe_nada],[freq_obs_ban, freq_espe_ban],[freq_obs_mu, freq_espe_mu])
```

```python
def grafica_de_residuos(*listas):
    """
    Función que grafica los residuos de múltiples ajustes de distribuciones.
    
    Parámetros:
    *listas: Tuplas o listas, cada una conteniendo dos listas [frecuencia_observada, frecuencia_esperada].
             Ejemplo:
             grafica_de_residuos(
                 [freq_obs_nada, freq_espe_nada],
                 [freq_obs_ban, freq_espe_ban],
                 [freq_obs_mu, freq_espe_mu]
             )
    """
    num_plots = len(listas)
    fig, axes = plt.subplots(1, num_plots, figsize=(6*num_plots, 5), squeeze=False)
    
    for idx, (y, ya) in enumerate(listas):
        residuos = [obs - esp for obs, esp in zip(y, ya)]
        ax = axes[0, idx]
        ax.plot(residuos, 'o', label='Residuos')
        ax.axhline(0, color='red', linestyle='--')
        ax.set_xlabel('Conteos')
        ax.set_ylabel('Residuos')
        ax.set_title(f'Residuos del Ajuste {idx+1}')
        ax.legend()
        ax.grid(True)
    
    plt.tight_layout()
    plt.show()
```

```python

def grafica_de_residuos(*listas):
    """
    Función que grafica los residuos de múltiples ajustes de distribuciones.
    
    Parámetros:
    *listas: Tuplas o listas, cada una conteniendo dos listas [frecuencia_observada, frecuencia_esperada].
             Ejemplo:
             grafica_de_residuos(
                 [freq_obs_nada, freq_espe_nada],
                 [freq_obs_ban, freq_espe_ban],
                 [freq_obs_mu, freq_espe_mu]
             )
    """
    
    num_plots = len(listas)
    fig, axes = plt.subplots(1, num_plots, figsize=(6*num_plots, 5), squeeze=False, sharey=True)
    
    # Colores y marcadores para diferenciar las gráficas
    colores = plt.cm.tab10.colors  # Colores diferentes para cada subplot
    marcadores = ['o', 's', '^', 'D', 'v', 'P', '*', 'X', 'h', '+']  # Diferentes marcadores
    
    for idx, (y, ya) in enumerate(listas):
        residuos = [obs - esp for obs, esp in zip(y, ya)]
        ax = axes[0, idx]
        
        # Graficar los residuos
        ax.scatter(range(len(residuos)), residuos, 
                   color=colores[idx % len(colores)], 
                   marker=marcadores[idx % len(marcadores)],
                   edgecolor='k', alpha=0.7, label='Residuos', s=100)
        
        # Línea horizontal en y=0
        ax.axhline(0, color='red', linestyle='--', linewidth=1)
        
        # Etiquetas y título
        ax.set_xlabel('Conteos', fontsize=12)
        if idx == 0:
            ax.set_ylabel('Residuos', fontsize=12)
        ax.set_title(f'Residuos del Ajuste {idx+1}', fontsize=14)
        
        # Leyenda y ajustes de la cuadrícula
        ax.legend(fontsize=10)
        ax.grid(True, which='both', linestyle='--', linewidth=0.5)
        
        # Mejorar los límites del eje y para visualizar mejor los residuos
        max_residuo = max(abs(min(residuos)), abs(max(residuos)))
        ax.set_ylim(-max_residuo*1.2, max_residuo*1.2)
    
    plt.tight_layout()
    plt.show()
```
