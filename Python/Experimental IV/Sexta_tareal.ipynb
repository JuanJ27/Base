# Reporte del Experimento de Franck–Hertz con Neón

## Introducción

El experimento de Franck–Hertz, realizado inicialmente en 1914 por James Franck y Gustav Hertz, es uno de los experimentos fundamentales que contribuyó a establecer la teoría cuántica y la estructura atómica moderna. En este experimento se evidencia que los átomos absorben energía de forma cuantizada, es decir, en porciones discretas (cuantos), lo cual valida los postulados propuestos por Niels Bohr en su modelo atómico.

En la versión aplicada al Neón, se observa que, al incrementar el voltaje acelerador $U_2$, la corriente que llega al ánodo presenta picos y valles. Estos se deben a que los electrones colisionan inelásticamente con los átomos de Neón, transfiriendo energía para excitarlos. Sólo aquellos electrones que conserven una energía cinética suficiente (por encima de un umbral, típicamente 1.5 eV) logran alcanzar el ánodo, mientras que los que han perdido energía no contribuyen a la corriente medida. De esta forma, se puede correlacionar la diferencia de potencial entre los picos con la energía de excitación de los átomos de Neón.

## Marco Teórico

El fundamento teórico del experimento se basa en los siguientes puntos:

- **Naturaleza Cuantizada de la Energía:**  
  La absorción y emisión de energía por parte de los átomos ocurre en “cuantos” o paquetes discretos. Esto implica que la energía transferida en las colisiones electrón-átomo no es continua, sino que se produce en saltos definidos.

- **Colisiones Inelásticas:**  
  Cuando un electrón colisiona inelásticamente con un átomo de Neón, puede transferir una cantidad específica de energía que corresponde a la diferencia entre dos niveles energéticos del átomo. Tras la colisión, el átomo queda en un estado excitado y posteriormente emite un fotón al volver a su estado fundamental o a un nivel inferior. En el caso del Neón, se han observado transiciones que involucran estados como los 3p y 3s, con diferencias energéticas de aproximadamente 18–19 eV entre picos y valores intermedios para las transiciones internas.

- **Medición de la Corriente y Detección de Picos:**  
  Al variar el voltaje acelerador $U_2$, se registra la corriente $I_c$ que llega al ánodo. La presencia de picos en la gráfica $I_c$ versus $U_2$ se interpreta como el resultado de los electrones que han colisionado y transferido energía en forma cuantizada. La diferencia de voltaje entre picos consecutivos se asocia a la energía necesaria para excitar a los átomos de Neón.

- **Relación Energía–Longitud de Onda:**  
  Utilizando la relación $E = \frac{hc}{\lambda}$, es posible calcular la longitud de onda del fotón emitido durante la transición inversa. En algunos experimentos (por ejemplo, en el caso del Mercurio se observa una emisión de 253.6 nm), pero para el Neón la radiación visible suele corresponder a transiciones que se ubican en la región roja o anaranjada.

## Resultados Esperados

En el desarrollo del experimento se espera obtener los siguientes resultados:

1. **Curva $I_c$ vs $U_2$ con Múltiples Picos:**  
   - Al aumentar gradualmente $U_2$, se debe observar una gráfica en la que la corriente $I_c$ presenta varios picos o franjas discretas.  
   - Se esperan picos espaciados aproximadamente por 18–19 V, lo que corresponde a la energía de excitación de los átomos de Neón.

2. **Detección de Picos y Cálculo de Diferencias de Potencial:**  
   - Mediante la detección de los picos en la curva experimental, se calcularán las diferencias de potencial ($\Delta U$) entre ellos.  
   - Estas diferencias se deben comparar con los valores reportados en la literatura para validar la cuantización de la energía en el Neón.

3. **Observación de Emisión Lumínica:**  
   - Durante el recorrido de los electrones a través del tubo, se debe observar la emisión de luz en la región entre la rejilla y el cátodo, asociada a la relajación de los átomos excitados.  
   - Se espera que la longitud de onda de la luz emitida, obtenida a partir de $E = hc/\lambda$, coincida (o se relacione) con la transición interna del Neón (por ejemplo, de 3p a 3s).

4. **Comparación de Mediciones con Diferentes Configuraciones:**  
   - Se pueden realizar mediciones con diferentes ajustes del sistema (por ejemplo, variaciones en el voltaje de frenado $U_3$ y en el voltaje de emisión $U_1$) para analizar la reproducibilidad y la influencia de los parámetros experimentales en la calidad de la señal obtenida.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, savgol_filter

# ==============================
# Datos Medidos
# ==============================
V = np.array([0.0, 1.1, 2.0, 2.9, 3.9, 4.9, 5.8, 7.0, 8.1, 9.0, 10.0, 11.1, 12.1,
              13.0, 14.1, 15.1, 16.1, 16.9, 18.1, 19.1, 20.1, 21.1, 22.0, 23.1, 24.0,
              25.5, 26.2, 27.1, 28.2, 29.0, 30.2, 31.2, 31.9, 33.1, 33.9, 34.9, 36.2,
              36.9, 38.2, 39.0, 40.0, 41.2, 43.1, 43.9, 45.2, 46.0, 47.2, 48.0, 49.0,
              50.0, 51.0, 52.1, 53.0, 54.2, 55.0, 56.0, 57.0, 58.0, 59.0, 60.0, 61.2,
              62.1, 63.0, 64.0, 65.1, 66.1, 67.0, 68.0, 69.0, 70.0, 71.1, 72.3, 72.9, 74.])
I1 = np.array([-0.05, -0.05, -0.04, -0.03, -0.03, -0.02, -0.02,  0.  ,  0.  ,
               0.  , 0.01,  0.02,  0.03,  0.03,  0.04,  0.05,  0.06,  0.08,
               0.08,  0.1,  0.11,  0.12,  0.13,  0.14,  0.15,  0.16,  0.17,
               0.19,  0.2,  0.21,  0.22,  0.25,  0.26,  0.26,  0.28,  0.29,
               0.3,  0.32,  0.33,  0.33,  0.35,  0.37,  0.39,  0.41,  0.43,
               0.46,  0.49,  0.52,  0.55,  0.58,  0.62,  0.65,  0.68,  0.71,
               0.72,  0.73,  0.74,  0.75,  0.76,  0.77,  0.79,  0.81,  0.83,
               0.86,  0.9,  0.93,  0.97,  1.03,  1.09,  1.13,  1.18,  1.22,
               1.24,  1.25])

error_V = 0.1  # error en V
error_I = 0.01  # error en I

# ==============================
# Suavizado y derivadas con propagación de error (aproximado)
# ==============================
ventana = 11  # entero impar
orden = 3

# Suavizado: asumimos que el error de I1 suavizado sigue siendo error_I.
I1_suave = savgol_filter(I1, window_length=ventana, polyorder=orden)
error_I1_suave = np.full_like(I1_suave, error_I)

# Para derivadas:
delta = error_V  # intervalo aproximado
dIdV = savgol_filter(I1, window_length=ventana, polyorder=orden, deriv=1, delta=delta)

# ==============================
# Detección de picos en dI/dV
# ==============================
indices_picos, propiedades = find_peaks(dIdV, prominence=0.005)
V_picos = V[indices_picos][-3:] if indices_picos.size else np.array([])

# ==============================
# Gráficas
# ==============================
plt.figure(figsize=(14, 7))

plt.subplot(2, 1, 1)
plt.grid(True)
plt.errorbar(V, I1, yerr=error_I, xerr=error_V, fmt='.', capsize=5,
             label='Datos Medidos', color='navy', markersize=4)
plt.plot(V, I1_suave, 'r-', linewidth=2, label='Ajuste Suavizado')
for v in V_picos:
    plt.axvline(v, color='red', linestyle='--', linewidth=1.5)
plt.xlabel('Voltaje U2 (V)', fontsize=14)
plt.ylabel('Corriente I (mA)', fontsize=14)
plt.title('Corriente vs Voltaje en el Experimento de Franck–Hertz (Neón)', fontsize=16)
plt.legend(fontsize=12)
plt.ylim(I1[0]-0.05, I1[-1]*1.2)
plt.xlim(V[0], V[-1])

plt.subplot(2, 1, 2)
plt.grid(True)
plt.plot(V, dIdV, 'g-', linewidth=2, label='Primera Derivada dI/dV')
for v in V_picos:
    plt.axvline(v, color='red', linestyle='--', linewidth=1.5)
plt.xlabel('Voltaje U2 (V)', fontsize=14)
plt.ylabel('dI/dV', fontsize=14)
plt.title('Primera Derivada de la Corriente vs Voltaje', fontsize=16)
plt.legend(fontsize=12)
plt.xlim(V[0], V[-1])

plt.tight_layout()
plt.show()

# ==============================
# Impresión de resultados con 2 cifras significativas
# ==============================
if V_picos.size:
    print("Voltajes (V) donde se detectan picos en dI/dV:",
          ", ".join(format(v, ".2g") for v in V_picos))
else:
    print("No se detectaron picos en dI/dV.")

if V_picos.size > 1:
    delta_V = np.diff(V_picos)  # diferencias entre picos
    error_delta_V_pair = np.sqrt(2)*error_V
    error_energia_excitacion = error_delta_V_pair / np.sqrt(delta_V.size)
    energia_excitacion = np.mean(delta_V)
    print("Diferencias entre picos (V):",
          ", ".join(format(dv, ".2g") for dv in delta_V),
          "±", format(error_delta_V_pair, ".2g"))
    print("Energía de excitación estimada: {} ± {} eV".format(
          format(energia_excitacion, ".2g"), format(error_energia_excitacion, ".2g")))
else:
    print("No se detectaron suficientes picos para determinar la energía de excitación.")
    energia_excitacion = None
    error_energia_excitacion = None

h_ev_s = 4.135667696e-15  # Constante de Planck en eV·s
c = 299792458  # Velocidad de la luz en m/s
if energia_excitacion and energia_excitacion > 0:
    lambda_nm = (h_ev_s * c / energia_excitacion) * 1e9  # nm
    error_lambda_nm = (h_ev_s * c * 1e9 / energia_excitacion**2) * error_energia_excitacion
    print("Longitud de onda correspondiente a la energía de excitación: {} ± {} nm".format(
         format(lambda_nm, ".2g"), format(error_lambda_nm, ".2g")))
else:
    print("No se pudo calcular la longitud de onda debido a la ausencia de picos o energía nula.")
```

## Conclusiones

El experimento de Franck–Hertz aplicado al Neón es una herramienta poderosa para demostrar la existencia de niveles de energía discretos en los átomos y la naturaleza cuantizada de la absorción de energía. A partir de la medición de la corriente en función del voltaje acelerador, se pueden identificar picos que reflejan las colisiones inelásticas de los electrones con los átomos, permitiendo así estimar la energía de excitación (aproximadamente 18–19 eV en el caso del Neón).

Entre las conclusiones que se pueden extraer del experimento destacan:

- **Validación de la Teoría Cuántica:**  
  La existencia de picos espaciados de manera regular confirma que los átomos de Neón sólo pueden absorber energía en cantidades discretas, apoyando los postulados del modelo de Bohr.

- **Importancia del Control Experimental:**  
  Los ajustes precisos de los voltajes $U_1$, $U_2$ y $U_3$ son fundamentales para la obtención de datos claros y reproducibles. Las variaciones en la incertidumbre y en el ruido de las mediciones afectan la visibilidad de los picos y, por ende, la precisión en la determinación de la energía de excitación.

- **Relación Energía–Emisión de Fotones:**  
  La energía transferida durante las colisiones inelásticas se relaciona con la longitud de onda de la luz emitida durante la relajación de los átomos excitados. Este vínculo es fundamental para comprender procesos como la fluorescencia y la emisión de luz en diferentes aplicaciones tecnológicas.

En resumen, el experimento no solo reafirma la cuantización de la energía en los átomos, sino que también ofrece una forma práctica de explorar y validar conceptos fundamentales de la física moderna.
