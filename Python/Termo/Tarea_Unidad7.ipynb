## Ecuaciones para la presion atmosferica

La troposfera es la capa de la atmósfera de un planeta que está en contacto directo con su superficie. Según Sánchez-Lavega y colaboradores *(Sánchez-Lavega et al., "Clouds in planetary atmospheres: a useful application of the Clausius-Clapeyron equation", American Journal of Physics 72 (6) (2004) 767-774)*, la relación entre la presión (P) y la altura (z) varía según el tipo de atmósfera. Para una atmósfera isotérmica, la relación se describe con la ecuación,
### Atmosfera isotermica
$$ P(z) = P_{0}e^{\frac{-gz}{RT_{0}}} $$

Donde $P_{0}$ es la presion a $z=0$, $T_{0}$ es la temperatura a $z=0$, g es la aceleración de la gravedad en el planeta y  es la constante específica del gas (la constante universal de los gases dividida por el peso molecular medio de la atmósfera).

En una atmósfera adiabática,
### Atmosfera adiabatica
$$ P(z) = P_{0} [ 1 - \frac{T_{a}(z-z_{0})}{T_{0}}]^{\frac{g}{RT_{a}}} $$
la relación se representa por la ecuación​ donde es el gradiente adiabático. Para comprender mejor estas relaciones, es necesario elaborar diagramas de las curvas para ambos tipos de atmósferas en al menos tres planetas y analizarlas.





## Consideraciones adicionales

Ademas de tener en cuenta las contantes relacionadas a la compocision quimica, debemos tener en cuenta los tamaños de los planteas debido a que al ser planetAS distintos, los valores de $z$ varian entre las graficas de cada uno.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the planets
planets = {
    'Venus': {'P0': 9.2e6, 'T0': 731, 'g': 8.89, 'R': 190, 'Gamma': 0.01050, 'z0':6052, 'z_max': 14000},
    'Marte': {'P0': 7e3, 'T0': 214, 'g': 3.74, 'R': 190, 'Gamma': 0.00450, 'z0':3400, 'z_max': 14000*0.5},
    'Jupiter': {'P0': 1e5, 'T0': 165, 'g': 24.25, 'R': 3750, 'Gamma': 0.00200, 'z0':69900, 'z_max': 14000*(69900/6052)},
}

# Plot the graphs
for planet, params in planets.items():
    P0 = params['P0']
    T0 = params['T0']
    g = params['g']
    R = params['R']
    Gamma = params['Gamma']
    z0 = params['z0']
    z_max = params['z_max']
    
    
    # Define the altitudes
    z_max_adiabatic = T0 / Gamma  # Maximum altitude for adiabatic atmosphere
    z = np.linspace(0, z_max, 1000)  # from 0 to 50 km
    #z_adiabatic = np.linspace(0, z_max_adiabatic, 1000)  # up to maximum valid altitude
    
    # Isothermal atmosphere
    P_iso = P0 * np.exp(-g * z / (R * T0))
    
    # Adiabatic atmosphere
    P_adiabatic = P0 * (1 - (Gamma * (z - z0) / T0)) ** (g / (R * Gamma))
    
    plt.figure()
    plt.plot(z, P_iso, label='Isotermal')
    plt.plot(z, P_adiabatic, label='Adiabática')
    plt.xlabel('Altitud (m)')
    plt.ylabel('Presión (Pa)')
    plt.title(f'Planeta {planet}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

## Analisis
En los tres planetas podemos apreciar que la presion ejercida de forma adiabatica de mantiene mayor o igual a la isotermica, siendo el igual unicamente en el caso de Jupiter para valores de $z$ mayores a 120$km$. Esto se puede deber a que precisamente jupiter cuenta con un valor de $R^{*}$ mucho mayor a los de venus y marte. Tambien interviene el hecho de que el dominio del $z$ para JUpiter es mucho mas extenso y le da pie a las funciones a encontrarse. Porque a nivel general en todas la graficas apreciamos que si bien las curvas son distintas, en el caso de marte y venus podemos intuir que con mas valores para $z$ hay un punto en el que ambas coinciden.

Como detalle adicionaly para finalizar, vemos claramente un comportamiento esperado y es que ambos tipos de presion disminuyen conforme aumenta la altura, lo cual tiene sentido ya que cuando aumentamos la altura, la columna de gas que ejerce la presion disminuye en volumen y peso.
