Escriba un código en Python para desarrollar el ejercicio 6.35. La cantidad n
moles de un gas ideal, evoluciona reversible e isobáricamente desde una
temperatura inicial Ti hasta una temperatura final Tf.
Calcúlense las variaciones de la energía interna, entalpía, entropía, potencial de
Helmholtz y potencial de Gibbs, todas estas variables termodinámicas en función
de Ti y Tf. Para un Ti dado, realice diagramas
de cada una de estas variables termodinámicas en función de la temperatura Tf
y compárelas. Realice los cálculos para cada uno de los gases que componen la
atmosfera terrestre asumiendo que son ideales y que experimentan variaciones normales de temperatura entre la
noche y el día e interprete los resultados.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
R = 8.314  # Constante de los gases ideales (J/(mol·K))

# Funciones para calcular propiedades termodinámicas

def energia_interna(n, Ti, Tf):
    return n * R * (Tf - Ti)

def entalpia(n, Ti,Tf):
    return n*R*(2*Tf-Ti)

def entropia(n, Ti, Tf):
    return n * R * np.log(Tf / Ti)

def potencial_helmholtz(n, Ti, Tf):
    return n*R*(Tf*(1-np.log(Tf/Ti))-Ti)

def potencial_gibbs(n, Ti, Tf):
    return n*R*(Tf*(2-np.log(Tf/Ti))-Ti)

# Parámetros
Ti = 273.15  # Temperatura inicial en Kelvin (0 °C)
n = 1.0      # Cantidad de moles del gas

# Rango de temperaturas finales Tf
Tf_range = np.linspace(Ti, 400, 100)  # Cambiar el rango según sea necesario

# Cálculo de las propiedades para cada Tf en el rango
energia_interna_vals = energia_interna(n, Ti, Tf_range)
entalpia_vals = entalpia(n, Ti, Tf_range)
entropia_vals = entropia(n, Ti, Tf_range)
potencial_helmholtz_vals = potencial_helmholtz(n, Ti, Tf_range)
potencial_gibbs_vals = potencial_gibbs(n, Ti, Tf_range)

# Graficar las propiedades en función de Tf y lo hacemos en un array de subplots
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

# Energía interna
axs[0, 0].plot(Tf_range, energia_interna_vals)
axs[0, 0].set_title('Energía interna')
axs[0, 0].set_xlabel('Tf (K)')
axs[0, 0].set_ylabel('U (J)')
axs[0, 0].grid()

# Entalpía
axs[0, 1].plot(Tf_range, entalpia_vals)
axs[0, 1].set_title('Entalpía')
axs[0, 1].set_xlabel('Tf (K)')
axs[0, 1].set_ylabel('H (J)')
axs[0, 1].grid()

# Entropía
axs[1, 0].plot(Tf_range, entropia_vals)
axs[1, 0].set_title('Entropía')
axs[1, 0].set_xlabel('Tf (K)')
axs[1, 0].set_ylabel('S (J/K)')
axs[1, 0].grid()

# Potencial de Helmholtz
axs[1, 1].plot(Tf_range, potencial_helmholtz_vals)
axs[1, 1].set_title('Potencial de Helmholtz')
axs[1, 1].set_xlabel('Tf (K)')
axs[1, 1].set_ylabel('A (J)')
axs[1, 1].grid()

# Potencial de Gibbs
axs[2, 0].plot(Tf_range, potencial_gibbs_vals)
axs[2, 0].set_title('Potencial de Gibbs')
axs[2, 0].set_xlabel('Tf (K)')
axs[2, 0].set_ylabel('G (J)')
axs[2, 0].grid()

# Temperatura usada
axs[2, 1].plot(Tf_range)
axs[2, 1].set_title('Temperatura')
axs[2, 1].set_xlabel('Tiempo')
axs[2, 1].set_ylabel('T (K)')
axs[2, 1].grid()

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

Este código define funciones para calcular cada una de las propiedades termodinámicas (energía interna, entalpía, entropía, potencial de Helmholtz y potencial de Gibbs) en función de la cantidad de moles $n$, la temperatura inicial $T_{i}$​  y la temperatura final $T_{f}$.

Finalmente, se utiliza matplotlib para graficar estas propiedades en función de $T_{f}$. Podemos tomar valores de $T_{i}$ y $T_{f}$ del tal forma que se asemeje al ciclo de un dia, tambien vemos que todas estas propiedades depende forma linear de el numero de moles, por lo que para simplificar el analisis podemos asumir que el numero de moles es constante.

```python
# Vamos a simular estos datos con una función seno de tal forma que el valor mas bajo de temperatura sea 17°C y el mas alto 28°C
# Generamos un array de temperaturas
sample = np.linspace(17, 24, 100)
# Generamos un array de valores de la función seno
Tf_range = (7*np.sin(sample) + 23)+273.15
# Graficamos la función de temperatura que imita el ciclo en un dia de forma idealizada
plt.plot(Tf_range)
plt.ylabel('Temperatura (K)')
plt.title('Variación de la temperatura en un día')
plt.grid()
plt.show()
```
