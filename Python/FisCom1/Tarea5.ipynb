Tarea sobre una caminata aleatoria en python, en clase vimos un ejemplo donde habian 4 direcciones, cada una con la misma probabilidad de ser elegida, en esta tarea se desarrolla una caminata donde se cuenta con 360° de direcciones posibles, pero descritas con una dstribucion normal, es decir, la mayoria de las direcciones estan cerca del angulo 0°, y conforme nos alejamos de este angulo, la probabilidad de elegir esa direccion disminuye. Habiendo algunos casos donde el angulo puede ser cercano a 180° o -180° ó el "caminante" se "devuelve".

```python
# Vamos a plotear un random walk, donde usamos dos variables, una radial para la distancia y otra angular para la dirección.
# para hacerlo un poco mas reaista, vamos a usar una distribucion normal para la distancia y la dirreccion.
# De tal forma que la distancia promedio sea 0.5 y desviacion de 0.1. Y el angulo de la direccion promedio sea 0 y desviacion de 30 grados
import numpy as np
import matplotlib.pyplot as plt
import math
import random


# Parametros del random walk
num_steps = 500
mean_distance = 0.5
std_distance = 0.1
mean_angle = 0
std_angle = np.pi*3 / 4 
# Antes de plotear la trayectoria, vamos a plotear la distribucion de las distancias y angulos
# Para comodidad, que el eje x del angulo vaya de -pi a pi
angles = np.random.normal(mean_angle, std_angle, num_steps)
distances = np.random.normal(mean_distance, std_distance, num_steps)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(distances, density=True, alpha=0.6, color='g')
plt.title('Distribucion de Distancias')
plt.xlabel('Distancia (metros)')
plt.ylabel('Frecuencia')
plt.subplot(1, 2, 2)
plt.hist(angles, density=True, alpha=0.6, color='b')
plt.title('Distribucion de Angulos')
plt.xlabel('Angulo (radianes)')
plt.ylabel('Frecuencia')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], ['-π', '-π/2', '0', 'π/2', 'π'])
plt.xlim(-np.pi*1.2, np.pi*1.2)
plt.tight_layout()
plt.show()
```

## 500 pasos

```python
# Vamos a plotear la trayectoria del random walk
x, y = [0], [0]
for _ in range(num_steps):
    angle = random.gauss(mean_angle, std_angle)
    distance = max(0, random.gauss(mean_distance, std_distance))  # Aseguramos que la distancia no sea negativa
    new_x = x[-1] + distance * math.cos(angle)
    new_y = y[-1] + distance * math.sin(angle)
    x.append(new_x)
    y.append(new_y)

plt.figure(figsize=(11, 7))
# Vamos a resaltar el punto inicial y final
plt.plot(x[0], y[0], marker='o', markersize=8, color='green', label='Inicio')
plt.plot(x[-1], y[-1], marker='o', markersize=8, color='red', label='Fin')
plt.plot(x, y, marker='o', markersize=2, linewidth=1, color='blue', alpha=0.7)
plt.legend()
plt.title('Random Walk en 2D')
plt.xlabel('X (metros)')
plt.ylabel('Y (metros)')
plt.axis('equal')
plt.grid(True)
plt.show()
```
